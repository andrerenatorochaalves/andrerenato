# Ensaios de Caracterização

Os ensaios de caracterização são a base de qualquer investigação geotécnica.

## 1. Granulometria

A análise granulométrica determina a distribuição do tamanho das partículas do solo.

### Coeficientes importantes

**Coeficiente de uniformidade:**

$$C_u = \frac{D_{60}}{D_{10}}$$

**Coeficiente de curvatura:**

$$C_c = \frac{(D_{30})^2}{D_{10} \cdot D_{60}}$$

## 2. Limites de Atterberg

| Parâmetro | Símbolo | Descrição |
|-----------|---------|------------|
| Limite de Liquidez | $LL$ | Teor de umidade na transição do estado plástico para o líquido |
| Limite de Plasticidade | $LP$ | Teor de umidade na transição do estado semi-sólido para o plástico |
| Índice de Plasticidade | $IP$ | $IP = LL - LP$ |
