# Ensaios de Caracterização

Os ensaios de caracterização são a base de qualquer investigação geotécnica.  Neste post, apresento os principais ensaios e como interpretar seus resultados.

## 1. Granulometria

A análise granulométrica determina a distribuição do tamanho das partículas do solo.

### Coeficientes importantes

**Coeficiente de uniformidade:**

$$C_u = \frac{D_{60}}{D_{10}}$$

**Coeficiente de curvatura:**

$$C_c = \frac{(D_{30})^2}{D_{10} \cdot D_{60}}$$

Onde:
- $D_{10}$ = diâmetro correspondente a 10% passante
- $D_{30}$ = diâmetro correspondente a 30% passante
- $D_{60}$ = diâmetro correspondente a 60% passante

## 2. Exemplo prático

Vamos analisar uma curva granulométrica típica de um rejeito de minério de ferro:

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo - Curva granulométrica
diametro_mm = np.array([0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1.0, 2.0, 5.0])
passante = np.array([5, 15, 25, 50, 65, 85, 92, 97, 100])

# Gráfico
fig, ax = plt. subplots(figsize=(10, 6))
ax. semilogx(diametro_mm, passante, 'b-o', linewidth=2, markersize=8)
ax.set_xlabel('Diâmetro das partículas (mm)', fontsize=12)
ax.set_ylabel('Porcentagem passante (%)', fontsize=12)
ax.set_title('Curva Granulométrica - Rejeito de Minério de Ferro', fontsize=14)
ax.grid(True, which='both', linestyle='--', alpha=0.7)
ax.set_xlim([0.001, 10])
ax.set_ylim([0, 100])

# Linhas de referência
ax.axhline(y=10, color='r', linestyle=':', alpha=0.5)
ax.axhline(y=30, color='r', linestyle=':', alpha=0.5)
ax.axhline(y=60, color='r', linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()
```

## 3. Limites de Atterberg

Os limites de Atterberg definem os estados de consistência do solo.

| Parâmetro | Símbolo | Descrição |
|-----------|---------|------------|
| Limite de Liquidez | $LL$ | Teor de umidade na transição do estado plástico para o líquido |
| Limite de Plasticidade | $LP$ | Teor de umidade na transição do estado semi-sólido para o plástico |
| Índice de Plasticidade | $IP$ | $IP = LL - LP$ |

## 4. Índices físicos

Os principais índices físicos são:

**Índice de vazios:**

$$e = \frac{V_v}{V_s} = \frac{\gamma_s}{\gamma_d} - 1$$

**Porosidade:**

$$n = \frac{V_v}{V_t} = \frac{e}{1+e}$$

**Grau de saturação:**

$$S = \frac{V_w}{V_v} = \frac{w \cdot G_s}{e}$$
